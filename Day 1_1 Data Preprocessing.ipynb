```
data.drop(["PassengerId","Name","Ticket","Cabin"],axis=1,inplace=True)
```

```
data.head()
```

```
data.isnull().sum()
```

```
import numpy as np
mean_age = data['Age'].mean()
mean_age = np.round(mean_age,2)
mean_age
```

```
data['Age'].fillna(mean_age,inplace=True)
data.isnull().sum()
```

```
data.dropna(inplace=True)
data.isnull().sum()
```

```
data.head()
```

```
data['Sex'].unique()
```

```
data['Embarked'].unique()
```

```
data['Sex'] = data['Sex'].map({"male":0,"female":1})
data['Embarked'] = data['Embarked'].map({"S":0,"C":1,"Q":2})
data.head()
```

```
#### rename the columns
data.rename(columns={"Sex":"Gender","Fare":"Price"},inplace=True)
data.head()
```

```
### Sort The data
data.sort_values("Price",ascending=False,inplace=True)
data
```

```
### Print data of only male passengers
male = data[data['Gender']==0]
male
```

```
data[data['Pclass']>1]
```

```
#### add a new column
data.reset_index(inplace=True)
data
```

```
data.drop(['index'],axis=1,inplace=True)
data.head()
```

```
data.head()
```

```
data.iloc[2]['Age']
```

```
s = []
for i in range(0,len(data)):
  a = data.iloc[i]['Survived']
  if a==0:
    s.append("Died")
  else:
    s.append("Alive")
```

```
data["Died/Alive"] = s
data.head()
```

```
data.to_csv("data.csv")
```
